# Disegni di possiibli distribuzione in funzione della distanza dalla sorgente

Per adesso il diametro del piattello non Ã¨ noto

```python
import matplotlib.pyplot as plt
import numpy as np

diam = 37.5 - 10. # cm
dist = 12. # cm
radius = diam / 2.
```

```python
x = np.linspace(-radius, radius, num= 300)
theta = np.arctan2(x, dist)

def cos_distr(theta, dist = dist):
    return np.cos(theta)**4 / dist**2

```

```python
plt.plot(theta, cos_distr(theta))
```
