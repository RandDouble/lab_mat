# Voglio fare un Plot in 3D

Uso la libreria "Plotly" per fare il nuovo grafico in quanto più efficente rispetto a MatPlotLib su questa categoria di grafico.
Da utilizzare è differente, ma non tanto.

```python
import plotly.graph_objects as go
# import pandas as pd # Tentativo per aprire file matlab nr 1
import numpy as np
# import h5py # Tentativo per aprire file matlab nr 2
import scipy.io
```

Uso scipy per aprire i file matlab, ~~~che in realtà sono in formato hdf5~~~, sono file di Matlab versione 5, sono compressi in qualche modo, ma sotto sotto sono file qualunque, scipy ha la libreria che li può gestire.
Ancora una volta si vede il fatto che usano software antico, sigh.

```python
# df = h5py.File("./data/Profilometer_data_flattened/21_04_I/M1_flat.mat", 'r')
df = scipy.io.loadmat("./data/Profilometer_data_flattened/21_04_I/M1_flat.mat")
data= np.array(df['matrice'])
```

```python
fig = go.Figure(data = [go.Surface(z = data)])
fig.update_layout(
    title = 'Profilo',
    autosize = False,
    width=500, height=500,
    margin=dict(l=65, r=50, b=65, t=90)
)
```
