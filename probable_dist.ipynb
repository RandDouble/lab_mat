# Disegni di possibili distribuzione in funzione della distanza dalla sorgente

```python
import matplotlib.pyplot as plt
import numpy as np

diam = 37.5 - 10. # cm
dist = 12. # cm
radius = diam / 2.
```

```python
x = np.linspace(-radius, radius, num= 300)
theta = np.arctan2(x, dist)

def cos_distr(theta, dist = dist):
    return np.cos(theta)**4 / dist**2
```

```python
# Questa è la distribuzione angolare con la distanza del piattello che abbiamo appena fissato.
# Questo è per avere un'idea anche degli angoli che andiamo a considerare effettivamente

plt.plot(np.degrees(theta), cos_distr(theta))
plt.xlabel(r"$\theta (degree)$")
```

```python
# Così ho solo una distanza dal piattello, forse ci posso giocare un po' su questa distanza, 
# provo a modificare d in maniera da vedere differnti caratteristiche del sistema
distances = [dist - i for i in np.arange(-2., 2.5, .5)]
for j in distances:
    plt.plot(x, cos_distr(np.arctan2(x, j)), label=f"distanza = {j:.1f} cm")
plt.xlabel("distanza da centro")
plt.ylabel("Massa Depositata")
plt.title("Distribuzione Coseno per varie distanze tra crogiolo e piattello")
plt.legend(loc="right")
plt.tight_layout()
```

è chiaro dalla distribuzione sopra esposta, che seppur questa manchi delle costanti di normalizzazione, l'altezza nel quale possiamo muoverci sia totalmente irrilevante.
Nonostante siano favorite le distanze più piccole, perchè vanno ad aumentare le differenze presenti, tuttavia non hanno una così significativo impatto rispetto a 12.0 cm da giustificare un loro uso.

Infatti se grafichiamo il rapporto tra la distibuzione  a 10.0 cm e tra 12.0 cm otteniamo il seguente risultato:

```python
plt.plot(x, cos_distr(np.arctan2(x , 10.0)) / cos_distr(np.arctan2(x , 12.0)))
plt.title("Differenza relativa")
```
